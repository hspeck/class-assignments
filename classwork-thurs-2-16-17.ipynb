```python
def makeBirdDict(filename):
    #this file opens the data file from THomas et al and extracts species and families as a dictionary
    birdDict={}
    fobj=open(filename, "r", encoding="ISO-8859-15") #open the file
    all_records=fobj.readlines()[2:] #read the file
    for line in all_records: #split lines on comma
        record_elements=line.split(",")
        species =record_elements[0]
        family=record_elements[2]
        birdDict[species]=family #stick species and families into birdDict
    return birdDict
```

```python
thomas_birds=makeBirdDict("/home/eeb177-student/Desktop/eeb-177/class-assignments/formatted_nature.csv")

species = list(thomas_birds.keys())
species.sort()
species[0]

print("family of "+ species[0]+ " is "+thomas_birds[species[0]])
```

```python

def get_at_contents(dna, sig_figs=2):
    length=len(dna)
    a_count=dna.upper().count("A")
    t_count=dna.upper().count("T")
    at_content=(a_count+t_count)/length
    return(round(at_content, sig_figs))
```

```python
seq="ATTTGTAGATA"
get_at_contents(seq)
```
